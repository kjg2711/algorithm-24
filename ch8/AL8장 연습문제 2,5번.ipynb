2)  미국은 그림과 같이 Quarter(25센트), Dime (10센트) , Nickel(5센트) penny(1센트)란 동전을 사용하고 있다. 이 동전들을 이용해 거스름돈 68센트를 돌려주려고 한다. 탐욕적 기법의 알고리즘 8.1이 수행 되는 과정을 설명하라. 또한 이방법이 거스름돈 동전 수를 최소화 할 수 있는지를 설명하라.

- `Quarter (25센트) / 2개  => 18 센트`
- `Dime (10센트) / 1개 => 8센트`
- `Nickel (5센트) / 1개 => 3센트 `
- `Penny (1센트) / 3개 = > 0센트`

5) A=(60kg, 10만원), B= (100kg, 20만원), C= (120kg, 30만원)의 세 물건이 있을 때 탐욕적 기법의 분할 가능한 배낭 채우기 알고리즘으로 용량 50kg인 배낭을 채우는 최적의 방법을 구하라. 

- 배낭 채우기 알고리즘이 항상 최적해를 보장하려면 단위 무게당 가격이 높은 것을 배낭에 채우면 최적해를 보장한다. 그러기 때문에 A, B, C 물건들의 단위 무게당 가치를 구하면 된다. 그래서 계산을 해보면 `A는 kg 당 1667원` `B는 kg당 2000원` `C는 kg 당 2500원` 이므로 C를 50kg인 배낭에 채우면 12.5만원으로 C를 배낭에 채우는 것이 최적의 방법이다.
