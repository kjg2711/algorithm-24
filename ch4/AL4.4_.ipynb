2. 알코리즘 성명:

정점의 진입차수를 계산하고 먼저 모든 정점의 진입차수를 0으로 초기화하고, 그래프의 모든 간선을 이용해 정점의 진입차수를 계산 

inDeg에서 진입차수가 0인 모든 정점을 찾아 vlist를 만든다.

vlist가 공백이 아니면 정점 v를 꺼내고 화면에 출력

v의 모든 인접 정점 u에 대해 진입차수를 감소시키고, u의 집입차수가 0으로 줄어들면 vlist에 u를 추가한다.

```python
# 5. 코드
# 알고리즘 4.4 위상정렬(축소 정복 기법) 
def topological_sort(graph) :
    inDeg = {}
    for v in graph :
        inDeg[v] = 0 
    for v in graph :
        for u in graph[v]:
            inDeg[u] += 1
    
    vlist = []

    for v in graph : 
        if inDeg[v] == 0 :
            vlist.append(v)
            
    while vlist :
        v = vlist.pop()
        print(v, end= ' ')


        for u in graph[v] :
            inDeg[u] -= 1
            if inDeg[u] == 0:
                vlist.append(u)
```
