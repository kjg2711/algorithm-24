4. 코드 개요(함수, 입력변수, 출력변수):

함수: def tsp(graph):

입력변수: graph = { A, B, C, D }

출력변수: return minpath, mindist / 최단경로 최단거리

```python
# 5. 코드
import itertools # 순열

def tsp(graph):
    city = list(graph.keys())
    mindist = float('inf')  # 무한대로 초기화
    minpath = None
    
    for path in itertools.permutations(city[1:]): 
        # iterable에서 요소의 연속된 길이 r 순열을 반환합니다.
        dist = graph[city[0]][path[0]]  # 출발 도시에서 첫 번째 도시까지의 거리

        for i in range(len(path) - 1):
            dist += graph[path[i]][path[i+1]]  # 현재 도시에서 다음 도시까지의 거리
        dist += graph[path[-1]][city[0]]  # 마지막 도시에서 출발 도시까지의 거리
        
        if dist < mindist:
            mindist = dist
            minpath = [city[0]] + list(path) + [city[0]]  # 최소 경로 업데이트
    return minpath, mindist
```

```python
# 6. 테스트 코드 
# 그래프를 딕셔너리로 표현
graph = {
    'A': {'A': 0, 'B': 2, 'C': 9, 'D': 10},
    'B': {'A': 1, 'B': 0, 'C': 6, 'D': 4},
    'C': {'A': 15, 'B': 7, 'C': 0, 'D': 8},
    'D': {'A': 6, 'B': 3, 'C': 12, 'D': 0}
}

# 외판원 문제 해결
minpath, mindist = tsp(graph)
print("최단 경로:", minpath)
print("최단 거리:", mindist)
```
