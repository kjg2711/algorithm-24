##### 2) 알코리즘 성명:
- def lp_delete 해시 테이블에서 선형 탐색을 하여 키를 삭제
    - `key % M` 을 통해 해시값을 계산하고 해시값을 이용하여 `id` 위치를 찾는다.
    - 찾는 항목이 해시 테이블에 있는경우 삭제한다.
    - 삭제한 위치는 `None` 처리하지 않고 `-1` 로 처리하여 해당 위치가 삭제되었다는 것을 표시한다.

- 입력 변수 :  
    -`key` : 삭제할 데이터의 키 값

- 출력 변수 : 



```python
# 5. 코드
def lp_delete(key) :
    id = hashFn(key)
    count = M 
    while count>0 :
        if table[id] == None :return 
        if table[id] != -1 and table[id] == key :
            table[id] = -1
            return 
        id = (id + 1 + M ) % M
        count -= 1
```
