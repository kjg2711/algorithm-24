# 알고리즘 예제 5.4 

1. 문제정의: 

이진트리의 높이 : 이진트리가 주어졌을 때 이 트리의 높이를 구하는 것

2. 알코리즘 성명:

분할 정복 결합으로 트리를 왼쪽 서브 트리와 오른쪽 서브 트리로 나누고 왼쪽 서브 트리와 오른쪽 서브 트리를 각각 높이를 구하고 각 서브 트리 중 큰 값에 1을 더하고 반환한다. 만약 루트가 None 이면 공백 트리이므로 높이를 0을 반환한다.

3. 손으로 푼 예제 (이미지삽입):



4. 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
  - def calc_height(root) : 공백 트리가 아니면 두 서브 트리의 높이를 계산하고 두 결과를 비교해 더 큰 값에 1을 더한 값 반환
- 입력 변수 : 
  - `root` : 트리의 루트 노드를 가리키고 전체 트리를 대표
- 출력 변수 : 트리의 높이
  - 트리의 `hLeft` 왼쪽 트리의 높이와 `hRight` 오른쪽 트리의 높이에서 둘 중 더 높은 높이 + 1

```python
# 5. 코드
def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hright = calc_height(root.right)
    return max(hLeft, hright) +1 
```

```python
# 6. 테스트 코드 
class TNode:
    def __init__ (self, data, left, right) :
        self.data = data
        self.left = left
        self.right = right
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print( "트리의 높이 = ", calc_height(root))
```

8. 복잡도 분석:

입력의 크기는 전체 트리의 노드 개수 n이다

`max(hLeft, hright) +1` 을 기본연산으로 하면  T(n) = O(n) 이다.

`root is None` 을 기본 연산으로 하면 모든 노드를 검사 하므로 점근적 시간 복잡도는 O(n)이다.
