선형 조사법의 탐색 알고리즘 

##### 2) 알코리즘 성명:
- lp_search 해시 테이블에서 선형 탐색을 사용하여 키를 검색
    - `key % M` 을 통해 해시값을 계산하고 해시값을 이용하여 `id` 위치를 찾는다.
    - 찾는 항목이 해당 위치에 해시 테이블 없는 경우 `None`을 반환한다.
    - 찾는 항목이 해당 위치 있는 경우 해당 값을 반환 만약 없으면 다음 위치 검사
    - `count` 를 이용해 해시 테이블을 순회

    -`key` : 데이터의 키값

- 출력 변수 : 
    - `None` 찾는 항목이 해시 테이블에 없는 경우
    - `table[id]` 찾는항목이 해시테이블 있는 경우
