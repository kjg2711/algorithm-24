5. 다음은 삽입 정렬의 변형된 알고리즘이다.



(1) 이 알고리즘을 파이썬으로 구현하라.

```python
def insertion_sort2(A, n):
    for i in range(1, n):
        j = i - 1
        while j >= 0 and A[j] > A[j+1]:
            A[j], A[j+1] = A[j+1], A[j]
            j = j - 1
```

```python
# 테스트 코드 
A = [5, 6, 7, 6, 4, 2, 1, 9, 10]
n = len(A)
insertion_sort2(A, n)
print(A)  # 정렬된 리스트 출력
```

(2) 이 알고리즘을 효율은 어떨까? 알고리즘 4.3과 비교하여 설명하라.

: 이 알고리즘은 A[j], A[j+1] = A[j+1], A[j] 이 부분이 계속해서 항복을 교환 복사하지만 알고리즘 4.3은 맨 마지막에 제 위치를 복사한다. 그러므로 알고리즘 4.3이 더 효율적이다.
