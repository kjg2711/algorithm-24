팩토리얼(하향식 축소 정복 기법) 순환구조

2. 알코리즘 성명:

n이 1일 때 1을 반환한다.

n이 1이 아닌 경우 n을 재귀적으로 호출하여 (n * factorial_recur(n-1)) 계산하고 반환한다.

4. 코드 개요(함수, 입력변수, 출력변수):

함수 : factorial_recur(n) 

입력 변수 : n 

출력변수 : 1 / (n * factorial_recur(n-1)) 

```python
# 5. 코드
# 알고리즘 4.1 팩토리얼 (하향식 축소 정복 기법)
def factorial_recur(n) :
    if n == 1 :
        return 1
    else :
        return (n * factorial_recur(n-1))
    
```

```python
def factorial_recur(n) :
    if n == 1 :
        return 1
    else :
        return (n * factorial_recur(n-1))

# 6. 테스트 코드 
n = int(input("입력 : "))
print(factorial_recur(n))
```
