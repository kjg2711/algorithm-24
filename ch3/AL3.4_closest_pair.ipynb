함수: def distance(p1, p2) ,closest_pair(p)

```python
# 5. 코드

# 알고리즘 3.4 최근접 쌍의 거리 
import math                  # math 모듈 포함. 제곱근(sqrt) 함수 사용을 위해 
def distance(p1, p2):        # 유클리드 거리 계산 함수
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)   # math.sqrt() 제곱근 계산 

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1 , n):
            dist = distance(p[i], p[j]) # 유클리드 거리 계산
            if dist < mindist:
                mindist = dist
    return mindist
```

```python
# 알고리즘 3.4 최근접 쌍의 거리 
import math                  # math 모듈 포함. 제곱근(sqrt) 함수 사용을 위해 
def distance(p1, p2):        # 유클리드 거리 계산 함수
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)   # math.sqrt() 제곱근 계산 

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1 , n):
            dist = distance(p[i], p[j]) # 유클리드 거리 계산
            if dist < mindist:
                mindist = dist
    return mindist

# 6. 테스트 코드
p = [(10,20), (30,40), (50,60), (1,2), (3,4), (5,6)]
print("최근접 거리: ", closest_pair(p))
```
