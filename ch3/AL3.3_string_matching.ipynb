1. 문제정의:

길이가 n인 입력 문자열 T와 길이가 m인 패턴 문자열 P가 있다. T에서 가장 먼저 나타나는 P의 위치를 찾아라. 패턴이 없으면 -1을 반환하라.

2. 알코리즘 성명:

억지 기법이며 텍스트의 첫 번쨰 문자 위치에 패턴을 놓고 비교한다. 이 과정을 패턴을 오른쪽으로 한칸씩 옮기면서 성공한 매칭이 나타날 때까지 반복한다.

4. 코드 개요(함수, 입력변수, 출력변수):

함수: def string_matching(T,P): T: 문자열 입력 / P: 문자열 패턴

입력변수: T 입력 , P 패턴 / text = input(str("문자열: ")), pattern = input(str("패턴 문자열: "))

출력변수: return i, return -1 / 문자열이 일치하면 해당위치 반환, 일치하지 않으면 -1 반환

```python
# 5. 코드

# 알고리즘 3.3 문자열 매칭(억지기법)
def string_matching(T,P): # T는 입력 P는 패턴 
    n = len(T)
    m = len(P)
    for i in range(n-m+1) :
        j = 0
        while j < m and P[j]== T[i+j] : # 패턴 문자열을 모두 비교
            j = j + 1
        if j == m :     # 모든 문자가 일치하면, 매칭 성공
            return i    # 현재 위치 반환 
    return -1           # 문자열 매칭 실패 
```

8. 복잡도 분석:

최선의 경우: 

첫워치에서 문자열이 패턴과 일치하는 경우이며, 이 경우는 패턴의 길이 m번의 비교만으로 문자열 매칭이 종료되므로 시간 복잡도는 O(m)이다.

최악의 경우: 

각 위치에서 패턴의 모든 묹라를 비교해야하는 경우이며, 복잡도 함수는 m(n-m+1) 이고, 시간 복잡도는 O(nm)이다.
