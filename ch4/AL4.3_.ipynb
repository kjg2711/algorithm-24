```python
# 5. 코드
# 알고리즘 4.3 삽입 정렬
def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = key  # 올바른 인덱스에 키를 삽입
        printStep(A, i) 

def printStep(A, i):  # 중간 과정 출력용 함수
    print(" Step %2d = " % i, end=' ')  # 현재 단계를 출력 
    print(A)  # 리스트의 내용을 출력 
```

```python
# 알고리즘 4.3 삽입 정렬
def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = key  # 올바른 인덱스에 키를 삽입
        printStep(A, i) 

def printStep(A, i):  # 중간 과정 출력용 함수
    print(" Step %2d = " % i, end=' ')  # 현재 단계를 출력 
    print(A)  # 리스트의 내용을 출력 

# 6. 테스트 코드 
data = [9, 8, 7, 6, 5, 4, 3, 2, 1]
print("Original:", data)
insertion_sort(data)
print("Insertion:", data)
```
