#### 알고리즘 6.3

##### 1) 문제정의: 

호스풀 알고리즘의 시프트 테이블 만들기

##### 2) 알코리즘 성명:
- 이 알고리즘은 문자열 매칭하는 호스풀 알고리즘의 시프트 테이블을 생성하는 함수 
    - `NO_OF_CHARS = 128 ` : ASCII 문자 기준인 리스트를 만든다.
    - `tbl = [m]*NO_OF_CHARS` :  시프트 테이블을 모든 항목을 m으로 채운다.
    - `for i in range(m-1)` : 패턴의 모든 문자에 대해 
    - `tbl[ord(pat[i])] = m -1 -i` : 그 알파벳이 패턴의 몇번째 문자인지 계산하여 시프트 테이블에 저장
    - `return tbl` :  시프트 테이블 반환

##### 3) 손으로 푼 예제 (이미지삽입):



##### 4) 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def shift_table(pat) :` 시프트 테이블를 생성하고 패턴의 위치에 따라 이동 거리 계산

- 입력 변수 :  
    - ` pat ` : 패턴문자열
    
- 출력 변수 : 
    - `tbl`: 테이블의 해당 위치의 이동 거리 (계산: `m -1 -i`)
 

```python
# 5. 코드
NO_OF_CHARS = 128
def shift_table(pat) :
    m = len(pat)
    tbl = [m]*NO_OF_CHARS

    for i in range(m-1) :
        tbl[ord(pat[i])] = m -1 -i

    return tbl
```

```python
# 6. 테스트 코드
def search_horspool(T,P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m-1 
    while( i<= n -1): 
        k = 0
        while k <= m -1 and P[m-1-k] == T[i-k] :
            k += 1
        if k == m :
            return i- m+1 
        else :
            tc = t[ord(T[i-k])]
            i += (tc-k)
    return -1

print("패턴의 위치: " , search_horspool("LIVERPOOLFC", "FC"))
```
