1. 문제정의:

리스트에 n개의 항목이 들어 있다. 이 리스트에서 "탐색키"를 가진 항목을 찾아라. 만약 찾는 항목이 있으면 그 항목의 인덱스를 반환하고, 없으면 -1을 반환하라. 단, 리스트의 항목들은 정렬되어있지 않다.

2. 알코리즘 성명:

순차 탐색의 예를 보여는 알고리즘이며 입력 리스트의 첫 항목부터 순서대로 하나씩 탐색키와 비교한다. 탐색에 성공하면 항목의 인덱스를 반환하고 찾는 항목이 없다면 -1을 반환한다.

4. 코드 개요(함수, 입력변수, 출력변수): 

함수 : sequential_search(A, key) / A는 입력 리스트 key는 탐색키

입력 변수 : A 리스트, key 탐색키 / A = [1,2,3,4,5], key = int(input("key: "))

출력 변수 : return i, return -1 / 탐색키로 찾을 때 인덱스 반환, 못 찾을 때 -1 반환

```python
# 5. 코드

# 알고리즘 3.2 순차 탐색(다시보기)
def sequential_search(A, key): # A는 입력 리스트, key는 탐색 키 
    for i in range(len(A)) :
        if A[i] == key : # 탐색 성공 (비교연산, 기본연산) 
            return i    # 탐색 성공 시 인덱스 반환
    return -1           # 탐색 실패 시 -1 반환


A = [1,2,3,4,5]
key = int(input("key: "))

print(sequential_search(A, key))
```

8. 복잡도 분석:

입력의 크기가 n이고 연산이(if A[i] == key :) == 이다. 그러므로 for문이 n번 반복 되기 때문에 O(n) 이다. 

최선의 경우: 리스트의 첫번째 항목이 key일 경우에는 O(1)

최악의 경우: 리스트 내에 Key가 없거나 맨 뒤에 있다면 O(n)

평균적인 경우 : (n+1)/2 O(n)
