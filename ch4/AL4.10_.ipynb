2. 알코리즘 성명:

리스트를 A.sort()를 통해 정렬하고 리스트에서 인덱스(k-1)을 반환

4. 코드 개요(함수, 입력변수, 출력변수):

함수 : 

        def kth_smallest_sort(A, K):

입력 변수 : 

        A / A = [1, 6, 7, 8, 9, 10,22,12,43,23,63,20]
        K / K = int(input("몇 번째: "))

출력 변수: 

        A[K - 1] : 인덱스

```python
# 5. 코드
# 알고리즘 4.10 정렬을 이용한 k번째 작은 수 찾기
def kth_smallest_sort(A, K):
    A.sort()
    return A[K - 1]
```

```python
# 5. 코드
# 알고리즘 4.10 정렬을 이용한 k번째 작은 수 찾기
def kth_smallest_sort(A, K):
    A.sort()
    return A[K - 1]

# 6. 테스트 코드 
A = [1, 6, 7, 8, 9, 10,22,12,43,23,63,20]
K = int(input("몇 번째: "))

print(kth_smallest_sort(A, K))
```
