## 알고리즘 예제 5.3

1. 문제 정의 

퀵 정렬

2. 알고리즘 설명 

- 정렬할 항목이 2개 이상인 경우에 대해서만 처리한다. 항목이 1개이면 이미 정렬된 것이다.

- 분할 알고리즘을 이용해 리스트를 좌우로 분할한다. 이때 리스트가 균등하게 분할된다는 보장은 없다. 분할이 완료되면 피벗의 인덱스가 반환되어 mid에 복사된다. 

- 왼쪽 부분 리스트 A[left..mid-1]와 오른쪽 부분 리스트 A[mid+1..right]를 각각 정렬한다. 

3. 손으로 푼 예제 :



4. 코드 개요(함수, 입력변수, 출력변수):
- 함수 : 
  - def quick_sort(A, left, right) : 퀵 정렬 리스트를 왼쪽 부분리스트와 오른쪽 부분 리스트로 나눠 퀵정렬한다.
  - partition(A, left, right) : 리스트를 좌우로 분할한다
- 입력 변수 : 
  - `A` : 리스트 , `left`,`right` : 정렬의 범위
- 출력 변수 : 트리의 높이
  - `quick_sort(data, 0, len(data)-1)` 리스트를 정렬 하고 병합 후 출력 

```python
# 5. 코드

def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]
    return high

def quick_sort(A, left, right) :
    if left<right :
        mid = partition(A, left, right) 
        quick_sort(A, left, mid-1)
        quick_sort(A, mid+1, right)
```

```python
# 6.테스트 코드

data = [ 1, 3, 4, 2, 5, 10, 8, 9, 7, 6 ]
print("Original : ", data)
quick_sort(data, 0, len(data)-1)
print("QuickSort :", data)
```

7. 수행 결과



QuickSort: 리스트를 좌우로 분할 하고 왼쪽과 오른쪽 부분리스트를 퀵 정렬후 병합 정렬 후 출력

8. 복잡도 분석

리스트 분할이 가운데에서 이루어진다. 항목의 수가 n이 2의 거듭제곱이라면 n/2^k의 크기로 나누어지게되므로 O(n log2 n) 이다.
