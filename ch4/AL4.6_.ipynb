1. 문제정의: 

이진 탐색(반복구조)

리스트에 n개의 항목이 들어있다. 이 리스트에서 "탐색키"를 가진 항목을 탐색하여 반환. 단, 리스트의 항목들은 정렬되어 있다.

2. 알코리즘 성명:

리스트의 중앙에 있는 항목을 탐색하고 이 항목이 탐색키보다 크면 찾는 항목은 왼쪽에 있고 오른쪽을 탐색할 필요가 없어진다. 반대로 그 항목이 탐색키보다 작다면 찾는 항목은 오른쪽에 있고 왼쪽은 탐색할 필요가 없어진다. 그래서 문제의 크기가 반으로 줄어든다. 그리고 탐색범위 내에서 탐색에 성공하면 찾는 항목의 인덱스를 반환한다

4. 코드 개요(함수, 입력변수, 출력변수):

함수 :   

        binary_search_iter(A, key, low, high) 

입력변수 : 

        A : 리스트 

        key : 찾는 항목 

        low,high : 탐색 범위 

출력변수 :

        -1 : 리스트 내에 찾는 항목(key값)이 없을 때 -1 반환 

        binary_search_iter(A, key, 0, len(A) - 1) : 리스트 내에 찾는 항목(key값)이 있을 때 인덱스 반환

```python
# 5. 코드
# 알고리즘 4.6 이진탐색 (반복구조) 
def binary_search_iter(A, key, low, high) :
    while (low <= high) :
        mid = (low + high) // 2 
        if key == A[mid]:
            return mid 
        elif key > A[mid]:
            low = mid + 1
        else : 
            high = mid - 1
    return -1 
```
