1. 문제정의: 

축소 정복 기법을 이용한 k번째 작은수 찾기 (리스트 분할)

2. 알코리즘 성명:

리스트 A의 left에서 right 사이의 항목을 분할 분할의 기준이 되는 항목의 인덱스가 반환

피벗의 인덱스 pos+1이 left+k와 같으면 전체리스트 A에서 left+k 번째 작은 항목을 찾은 것

만약 pos+1이 left+k보다 크면 답은 왼쪽 그렇지 않으면 오른쪽 부분 리스트에 있다. 

low와 high가 역전되지 않는 동안 항목의 크기가 피벗을 기준으로 반대인 위치까지 이동 low는 증가 high는 감소 방향 탐색

선택된 두 레코드 교환 

high와 피벗의 항목을 교환 high 반환

4. 코드 개요(함수, 입력변수, 출력변수):

함수 :

    def partition(A, left, right) :

입력변수: 

    array = [ ]

출력변수: 

    quick_select(array, 0 ,n-1, 1)


```python
# 5. 코드
# 알고리즘 4.11 축소 정복 기법을 이용한 K번째 작은수 찾기
def quick_select(A, left, right, k):
    pos = partition(A,left,right)
    
    if (pos+1 == left+k):
        return A[pos]
    elif (pos+1 > left+k):
        return quick_select(A,left,pos-1,k)
    else :
        return quick_select(A, pos+1, right, k-(pos+1-left)) 
    
# 알고리즘 4.12 리스트 분할(Hoare partition)
def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]
    return high
```

```python
# 6. 테스트 코드 
array = [18, 92 ,4, 23, 10, 20, 8, 9]
print("입력 리스트 =", array)
n = len(array)
print("[축소정복] 3번째 작은 수: ", quick_select(array, 0, n-1, 3))
```
