팩토리얼 (상향식 축소 정복기법) 반복구조

2. 알코리즘 성명:

n에 대해 1부터 n번 까지 반복문을 통해 반복하고 result를 연산하고 result를 반환

4. 코드 개요(함수, 입력변수, 출력변수):

함수: 

        def factorial_iter(n) :

입력변수 :

        n

출력변수 :

        result

```python
# 5. 코드
# 알고리즘 4.2 팩토리얼 (상향식 축소 정복 기법) 
def factorial_iter(n) :
    result = 1
    for k in range(1, n+1) :
        result = result * k
    return result 
    
```

```python
# 알고리즘 4.2 팩토리얼 (상향식 축소 정복 기법) 
def factorial_iter(n) :
    result = 1
    for k in range(1, n+1) :
        result = result * k
    return result 
    

# 6. 테스트 코드 
n = int(input("입력 : "))
print(factorial_iter(n))
```
