1) 문제정의: 

리스트에 n 개의 항목들이 들어 있다. 이들을 키(key)의 순서에 따라 오름차순(non-decreasing order)으로 재배치하라.

2) 알고리즘 설명: 

입력 리스트에서 가장 작은 항목을 찾고, 이것을 꺼내 정렬된 리스트에 순서대로 저장한다.

4) 코드 개요: 

함수 : selection_sort(A), printStep(arr, val) / 선택정렬 함수, 중간출력 함수 

연산 : (A[j] < A[least]) / < 

입력변수 : data = [9,8,7,6,5,4,3,2,1]

출력변수 : 

```python
# 5. 코드 

# 알고리즘 3.1 선택정렬
def selection_sort(A) : # 선택정렬 
    n =len(A)
    for i in range(n-1):
        least = i
        for j in range(i+1, n):
            if (A[j]<A[least]) : # 비교 연산 < 
                least = j       # 최소항목 갱신
        A[i], A[least] = A[least], A[i]
        printStep(A, i + 1)
        
def printStep(arr, val) :       # 중간 과정 출력용 함수
    print(" Step %2d = " % val, emd=' ') # 현재 단계를 출력 
    print(arr)          # 리스트의 내용을 출력 
```

```python
# 알고리즘 3.1 선택정렬
def selection_sort(A) : # 선택정렬 
    n =len(A)
    for i in range(n-1):
        least = i
        for j in range(i+1, n):
            if (A[j]<A[least]) : # 비교 연산 < 
                least = j       # 최소항목 갱신
        A[i], A[least] = A[least], A[i]
        printStep(A, i + 1)
        
def printStep(arr, val) :       # 중간 과정 출력용 함수
    print(" Step %2d = " % val, emd=' ') # 현재 단계를 출력 
    print(arr)          # 리스트의 내용을 출력 
    
# 6. 테스트 코드
data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original : ", data)
selection_sort(data)
print("Selection : ", data)
```
