##### 2) 알코리즘 성명:
버킷(큐)를 통하여 입력 항목들을 순서대로 킷값에 따라 해당 버킷에 넣고 순차적으로 버킷 안에 들어 있는 수를 정렬 하여 출력한다.

##### 4) 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def radix_sort()` 버킷을 이용하여 기수 정렬 


- 입력 변수 :  
    - `BUCKETS` 버킷의 수 (진법) , `DIGITS` 전체 패스의 수 (자릿 수) 
    - `queues` 큐의 리스트 : 버킷
    - `data` 숫자 리스트 (정렬할 데이터)


- 출력 변수 : 
    - `radix_sort(data)`: 기수 정렬 `print("Radix: ", data)`: 결과 출력 
 

```python
# 5. 코드
from queue import Queue
def radix_sort(A) :
    queues = [] 
    for i in range(BUCKETS) :
        queues.append(Queue())
    n = len(A)
    factor = 1
    for d in range(DIGITS) :
        for i in range(n) :
            queues[( A[i]//factor ) % BUCKETS].put(A[i])
        j = 0 
        for b in range(BUCKETS) :
            while not queues[b].empty() :
                A[j] = queues[b].get()
                j += 1
            factor *= BUCKETS
            print("step", d+1, A)
            
```

```python
# 6. 테스트 코드 
import random
BUCKETS = 10
DIGITS = 5
data = []
for i in range(10) :
    data.append(random.randint(1,99999))
radix_sort(data)
print("Radix: ", data)
```

##### 8) 복잡도 분석:
입력의 개수가 n이다 따라서 T(n) = d(n+n) 이고, 기수 정렬은 O(d*n)의 시간 복잡도를 갖는다.
