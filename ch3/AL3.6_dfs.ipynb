1. 문제정의:

깊이 우선 탐색을 이용하여 정점을 순회하여 방문하라

2. 알코리즘 성명:

탐색은 시작 정점에서 출발하여 인접 정점을 임의로 하나 선택하여 탐색하고 방문한다. 

이미 탐색한 정점은 다시 방문하지 않아야한다. 

방문하지 않은 인접 정점을 탐색하고 방문한다. 

방문하지 않은 인접 정점이 없다면, 이전 정점으로 되돌아가서 방문하지 않은 인접 정점을 살핀다. 가지 않은 정점이 있으면 탐색 방문한다. 

더 이상 방문하지 않은 인접 정점이 없으면 이전 정점으로 되돌아 온다. (시작 정점까지)

시작 정점에서도 더 이상 방문하지 않은 정점이 없다면 알고리즘은 종료된다.

4. 코드 개요(함수, 입력변수, 출력변수)

함수: dfs( graph, start, visited ) / graph 그래프 start 시작정점 visited 방문 집합

입력변수: mygraph = { }

출력변수: 

```python
# 5. 코드

# 알고리즘 3.6 깊이 우선 탐색 (순환)
def dfs( graph, start, visited ):   # 깊이 우선 탐색 알고리즘
    if start not in visited :
        visited.add(start)
        print(start, end='')
        nbr = graph[start] - visited 
        for v in nbr:
            dfs(graph, v, visited)
```
