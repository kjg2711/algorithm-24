##### 2) 알코리즘 성명
- 이 알고리즘은 경험적인 전략(휴리스틱)들을 이용 
    - 휴리스틱을 사용하려면 패턴에 대한 전처리 과정이 필요하다.
    - 모든 가능한 알파벳에 대해 패턴에서의 위치를 저장한 표(시프트 테이블)
    - 텍스트 문자열 T, 패턴 문자열 P 일치하는 부분을 찾는다.
    - 하나의 위치(shift)에서 패턴은 뒤에서 부터 앞으로 검사한다. 
    - 패턴 문자열 P와 일치하면 해당 인덱스 반환 , 일치하지 않으면 -1 반환

##### 4) 코드 개요(함수, 입력변수, 출력변수)
- 함수 : 
    - `def search_Horspool()` : T에서 P가 나타나는 위치를 찾아 반환 

- 입력 변수 :  
    - `T` : 텍스트(입력) 
    - `P` : 패턴 

- 출력 변수 : 
    - `print("패턴의 위치: ")` : 패턴의 위치
    - `i-m+1`: 매칭성공 : 매칭 위치(왼쪽) 반환
    - `-1 ` : 매칭실패 -1 반환

```python
# 5. 코드
def search_horspool(T,P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m-1 
    while( i<= n -1): 
        k = 0
        while k <= m -1 and P[m-1-k] == T[i-k] :
            k += 1
        if k == m :
            return i- m+1 
        else :
            tc = t[ord(T[i-k])]
            i += (tc-k)
    return -1
```

```python
# 6. 테스트 코드 
# 6.5 
NO_OF_CHARS = 128
def shift_table(pat) :
    m = len(pat)
    tbl = [m]*NO_OF_CHARS

    for i in range(m-1) :
        tbl[ord(pat[i])] = m -1 -i

    return tbl

print("패턴의 위치: " , search_horspool("APPLEMANGOBANANAGRAPE", "MANGO"))
```
