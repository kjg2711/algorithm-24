##### 1) 문제정의: 

선형 조사법의 삽입 알고리즘

##### 2) 알코리즘 성명:
- 삽입
    - 해시 함수로 킷값을 M으로 나눈 나머지 연산 사용 `key % M` 해시 값
    - 킷값을 해시 함수에 적용해 해시 주소 id를 구한다.
    - `while count>0 and (table[id] != None and table[id] != -1 )` id 테이블이 비어있지 않으면 
    - `id = (id + 1 + M )% M` 다음위치로 이동하여 검사
    - `table[id] = key` 항목(키)을 저장한다.

##### 4) 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def lp_insert(key):` 해시 테이블에 데이터를 삽입 연산 함수


- 입력 변수 :  
    - `M` : 테이블의 크기 : 소수를 사용한다.
    - `key` : 삽입할 데이터의 값

- 출력 변수 : 
 

```python
# 5. 코드
M = 13
table = [None]*M

def hashFn(key) :
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    # while count > 0 and (table[id] != None) :
    while count>0 and (table[id] != None and table[id] != -1 ) :
        id = (id + 1 + M )% M
        count -= 1
    if count > 0:
        table[id] = key 
    return 
```

```python
# 6. 테스트 코드 
#6.7
def lp_search(key) :
    id = hashFn(key)
    count = M 
    while count > 0:
        if table[id] == None :
            return None
        if table[id] != -1 and table[id] == key :
        # if table[id] == key :
            return table[id]
        id = (id+ 1 + M )% M
        count -= 1
    return None 

#6.8
def lp_delete(key) :
    id = hashFn(key)
    count = M 
    while count>0 :
        if table[id] == None :return 
        if table[id] != -1 and table[id] == key :
            table[id] = -1
            return 
        id = (id + 1 + M ) % M
        count -= 1
        
print(" 최초: ", table)
lp_insert(45); print("45 삽입", table)
lp_insert(30); print("30 삽입", table)
lp_insert(60); print("60 삽입", table)
lp_delete(60); print("60 삭제", table)
print("탐색:", lp_search(45))
```
