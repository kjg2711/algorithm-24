```python
# 5. 코드
# 4.7 알고리즘 거듭제곱(억지기법)
def slow_power(x, n):
    result = 1.0 
    for i in range(n):
        result = result * x
    return result 
```

```python
# 4.7 알고리즘 거듭제곱(억지기법)
def slow_power(x, n):
    result = 1.0 
    for i in range(n):
        result = result * x
    return result 
# 6. 테스트 코드 
# 밑 지수 입력
x = float(input("밑 입력: "))
n = int(input("지수 입력: "))
print(slow_power(x, n))


import time 
print(f" 억지기법({x},{n}) = ", slow_power(x,n))

t2 = time.time()
for i in range(100000) : slow_power(x,n)
t3 = time.time()
print("억지기법 시간 :", t3-t2) 
```
