12) 알파벳 소문자와 대문자(a~z, A~Z)만으로 이루어진 리스트를 알파벳 순으로 정렬하는 코드를 카운팅 정렬로 구현하라

```python
def counting_sort(A) :
    MAX_VAL =123  
    MIN_VAL =65   
    
    output = [0] *len(A)
    count = [0] * MAX_VAL
    
    print("Original : ", A)
    
    for char in A:
        count[ord(char)] +=1
    print("Counting : ", [count[ord(char)] for char in A])
    
    for i in range(MIN_VAL +1, MAX_VAL):
        count[i] += count[i -1]
    
    print("Counting : ", [count[ord(char)] for char in A])
    for i in range(len(A) -1, -1, -1):
        output[count[ord(A[i])] -1] = A[i]
        count[ord(A[i])] -=1
    for i in range(len(A)):
        A[i] = output[i]
```
