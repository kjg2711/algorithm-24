1. 문제정의:

너비 우선 탐색은 시작 정점으로부터 가까운 정점을 먼저 방문하고 멀리 떨어져 있는 정점을 나중에 방문한다.

2. 알코리즘 성명:

시작 정점을 정한다. 처음에는 큐에 시작 정점을 삽입해 해둔다.

큐에서 이전 정점을 꺼내고 아직 방문하지 않은 인접 정점을 순서대로 방문하고 이들을 큐에 넣는다. 

다시 큐에서 이전 정점을 꺼내고 방문하지 않은 이웃 정점을 방문하고 큐에 넣는다. 이것을 반복한다.

만약 큐에서 이전 정점을 꺼냈을 때 방문하지 않은 이웃정점이 없고 큐에서 연속으로 꺼내도 마찬가지이다.

그러므로 큐는 다시 공백 상태가 되고 탐색은 종료 된다.

함수: bfs( graph, start) / graph 그래프 딕셔너리 형태 start 시작정점 

입력변수: mygraph = { }

출력변수: 

```python
# 5. 코드
# 알고리즘 3.7 너비 우선 탐색 (반복)
import queue 
def bfs(graph, start):
    visited = { start }
    que = queue.Queue()
    que.put(start)
    while not que.empty():
        v = que.get()
        print(v, end='')
        nbr = graph[v] - visited 
        for u in nbr:
            visited.add(u)
            que.put(u)
```

```python
# 알고리즘 3.7 너비 우선 탐색 (반복)
import queue 
def bfs(graph, start):
    visited = { start }
    que = queue.Queue()
    que.put(start)
    while not que.empty():
        v = que.get()
        print(v, end='')
        nbr = graph[v] - visited 
        for u in nbr:
            visited.add(u)
            que.put(u)

# 6. 테스트 코드 
mygraph = { "A" : {"B","C"}, 
            "B" : {"A","D"},
            "C" : {"A","D","E"},
            "D" : {"B","C","F"},
            "E" : {"C","G","H"},
            "F" : {"D"},
            "G" : {"E","H"},
            "H" : {"E","G"}
            }

print('BFS :', end='')
bfs(mygraph,"A")
print()
```
