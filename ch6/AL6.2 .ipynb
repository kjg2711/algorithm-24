#### 알고리즘 6.2

##### 1) 문제정의: 

카운팅 정렬

##### 2) 알코리즘 성명:
- 카운팅 정렬
    - `output ` 정렬된 결과를 저장할 임시 리스트
    - `count` 각 숫자별 빈도를 계산하고 해당 숫자의 위치가 되도록 수정
    - 모든 입력 항목 `[count[A[i]]-1] = A[i] ` 에 저장 
    - ` A[i] = output[i]` 정렬 결과를 원래 배열에 복사

##### 4) 코드 개요(함수, 입력변수, 출력변수):

- 함수 : 
    - `def counting_sort(A): ` 입력된 리스트를 0부터 MAX_VAL-1 까지인 리스트 정렬

- 입력 변수 :  
    - `data` :  정렬할 리스트
    - `MAX_VAL` : 데이터의 최대값
    
- 출력 변수 : 
    - `print("Counting :", data)` : 정렬된 데이터(리스트)
 

```python
# 5. 코드
def counting_sort(A): 
    output= [0] * len(A) 
    count = [0] * MAX_VAL

    for i in A:
        count[i] += 1

    for i in range(MAX_VAL): 
        count[i] += count[i-1]
    
    for i in range(len(A)) :
        output[count[A[i]]-1] = A[i] 
        count[A[i]] -= 1
    for i in range(len(A)):
        A[i] = output[i] 

```

```python
# 6. 테스트 코드 
MAX_VAL = 10 
data = [8,  7 , 1, 3, 7, 5, 6 ] 
print("Original : " , data)
counting_sort(data)
print("Counting :", data) 
```

입력의 크기는 n이고 입력의 크기에 비례한다. 따라서 시간 복잡도는 O(n)이고 공간 복잡도는 입력 리스트와 출력 배열크기가 `n` 이므로 O(n)이다.
