# 알고리즘 예제 5.6

1. 문제정의: 

이진트리의 중위 순회 : 왼쪽 서브 트리 L, 트리의 각 노드를 루트 V, 오른쪽 서브 트리 R을 순서대로  순회한다.

2. 알코리즘 성명:

- 중위 순회 LVR의 순서 대로 순회한다.  
  - 왼쪽 서브 트리를 순회한다. `L` 
  - 루트노드를 방문하여 처리한다. `V`
  - 오른쪽 서브 트리를 순회한다. `R` 



3. 손으로 푼 예제 (이미지삽입):



if n is not None: : 종료 조건

4. 코드 개요(함수, 입력변수, 출력변수):

- 함수 :
  - def preorder() : 전위 순회로 VLR순으로 순회한다. 
- 입력변수 :
  - 노드의 전체 트리
- 출력변수 :
  - `print(n.data, end='')` 각 노드를 순회 순서로 출력 
print(n.data, end='')

```python
# 5. 코드
def inorder(n) :
    if n is not None :
        inorder(n.left)
        print(n.data, end='')
        inorder(n.right)
```

```python
# 6. 테스트 코드 
class TNode:
    def __init__ (self, data, left, right) :
        self.data = data
        self.left = left
        self.right = right
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order: ',end='')
inorder(root)
print()
```

8. 복잡도 분석:

입력의 크기는 전체 트리의 노드 개수 n이다.

`출력문 print(n.data, end=' ')`을 기본 연산으로 하면 복잡도 함수는 T(n) = n , O(n)이 된다.

`공백트리 검사 if n is not None: `을 기본연산으로 한다면 모든 노드에서 검사가 이루어져야 하므로 T(n)= n + x , O(n)이 된다.
