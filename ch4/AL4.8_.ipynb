2. 알코리즘 성명:

n = 0이면 1을 반환 

n이 짝수면 power(x * x, n // 2) 반환

n이 홀수면 x * power(x * x, (n - 1) // 2) 반환

```python
# 5. 코드
# 4.8 알고리즘 거듭제곱(축소 정복 기법)
def power(x, n):
    if n == 0 :
        return 1
    elif (n % 2) == 0 :
        return power(x * x, n // 2)
    else :
        return x * power(x * x, (n - 1) // 2)
```

```python
# 4.8 알고리즘 거듭제곱(축소 정복 기법)
def power(x, n):
    if n == 0 :
        return 1
    elif (n % 2) == 0 :
        return power(x * x, n // 2)
    else :
        return x * power(x * x, (n - 1) // 2)
    
# 6. 테스트 코드 
# 밑 지수 입력
x = float(input("밑 입력: "))
n = int(input("지수 입력: "))
print(power(x, n))


import time 
print(f"축소정복기법({x},{n}) = ", power(x,n))

t1 = time.time()
for i in range(100000) : power(x,n)
t2 = time.time()
print("축소정복기법 시간 :", t2-t1) 
```

8. 복잡도 분석:

n을 2의 거듭제곱 값의 하나인 2^k 라고 가정하고 순환호출을 한번 할 때 마다 n의 크기가 절반씩 줄어들게 된다. 그래서 k번의 순환 호출이 일어난다. 그러기 때문에 n = 2^k 이므로 양변에 log 함수를 취하면 log2n = k가 되므로 시간복잡도는 O(log2n)이 된다.
