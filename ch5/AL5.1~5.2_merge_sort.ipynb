2. 알코리즘 성명 :
- 병합 정렬 / 병합 알고리즘 (정렬된 두 리스트의 병합)
  - 각 부분 리스트의 첫번째 항목들부터(i=left, j=mid+1) 오른쪽으로 시작한다.
  - 두 부분 리스트의 항목 A[i]와 A[j]를 비교하여 더 작은 요소를 먼저 새로운 리스트 sorted로 복사한다.
  - A[i]>A[j] 같은 경우는 A[j]를 sorted[k]로 복사하고 인덱스 j와 임시 리스트의 인덱스 k를 증가 시킨다.
  - A[i] =< A[j] 같은 경우는 A[i]를 sorted[k]로 복사하고 i와 k를 증가 시킨다.
  - 다른 리스트의 남은 항목들을 모두 C로 복사한다.

4. 코드 개요(함수, 입력변수, 출력변수) :

- 함수 : 
  - def merge_sort(A, left, right) : / 병합 정렬 : 정렬할 항목이 2개 이상이면 분할 - 정복을 진행한다.
  - def merge(A, left, mid, right) : / 병합 알고리즘 (정렬된 두 리스트의 병합) : 정렬된 두 리스트를 병합한다. 둘을 병항해 하나의 정렬된 리스트를 만든다.

- 입력 변수 :
  - `A` : 입력 리스트, `left`, `right` : 리스트 시작 항목 과 끝 항목

- 출력변수 :
  - `print("Original : ", data)` : 최초의 리스트
  - `print("MergeSort : ", data)` : 병합 정렬 후 리스트 

```python
# 5. 코드
# 알고리즘 5.1 병합 정렬 (순환(재귀))
def merge_sort(A, left, right) :
    if left<right :
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid ,right)
# 5. 코드
# 알고리즘 5.2 병합 알고리즘(정렬된 두 리스트의 병합
def merge(A, left, mid, right) :
    k = left 
    i = left
    j = mid + 1
    while i<= mid and j<=right:
        if A[i] <= A[j] :
            sorted[k] = A[i]
            i, k = i+1, k+1
        else :
            sorted[k] = A[j]
            j, k = j+1, k+1
    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]

    A[left:right+1] = sorted[left:right+1]
```
